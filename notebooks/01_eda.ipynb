```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("data/raw/smart_meter.csv")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp")

plt.figure(figsize=(12,4))
plt.plot(df["timestamp"][:500], df["consumption_kwh"][:500])
plt.title("Energy Consumption (First 500 points)")
plt.xlabel("Time")
plt.ylabel("kWh")
plt.show()
```
